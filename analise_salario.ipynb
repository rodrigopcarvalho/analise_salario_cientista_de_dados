```python
# importando bibliotecas

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# leitura do data set

df = pd.read_csv('/kaggle/input/salary-data/Salary_Data.csv')
```

```python
# exibir as 5 primeiras linhas do data set

df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
# renomeando as colunas

df.columns = ['idade', 'genero', 'nivel_escolar', 'cargo', 'anos_experiencia', 'salario']
```

```python
df.head(1)
```

```python
df = df.replace({"Bachelor's Degree":"Bachelor",
                 "Bachelor's":"Bachelor",
                 "Master's Degree":"master",
                 "Master's":"master",
                 "PhD":"phD"})
```

```python
# averiguando quntidade de valores por coluna
df.nunique()
```

```python
# valores na coluna idade

df['idade'].value_counts()
```

```python
# valores na coluna genero

df['genero'].value_counts()
```

```python
# valores na coluna nivel_escolar

df['nivel_escolar'].value_counts()
```

```python
# valores na coluna cargo

df['cargo'].value_counts()
```

```python
# valores na coluna anos_experiencia

df['anos_experiencia'].value_counts()
```

```python
# valores na coluna salario

df['salario'].value_counts()
```

```python
# averiguando valores NaN

df.isna().any()
```

```python
# quantidade de valores NaN por coluna

df.isna().sum()
```

```python
# removendo valores NaN

df = df.dropna(how='any', axis=0)
```

```python
# quantidade de valores NaN por coluna

df.isna().sum()
```

```python
# salvando data set com os dados tratados

df.to_csv('dados_salario_tratado.csv', index=False)
```

```python
plt.figure(figsize=(5,3))

plt.bar(df['genero'], df['salario'], color='k')

plt.title('Relação Salário x Gênero')
plt.xlabel('genero')
plt.ylabel('salario')
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.bar(df['idade'], df['salario'])
plt.title('Relação Idade x Salário')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.bar(df['nivel_escolar'], df['salario'])
plt.title('Relação Nivel Escolar x Salário')
plt.xlabel('Nivel Escolar')
plt.ylabel('Salário')
plt.xticks(rotation=70)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.bar(df['anos_experiencia'], df['salario'])
plt.title('Relação Anos de Experiência x Salário')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.show()
```
